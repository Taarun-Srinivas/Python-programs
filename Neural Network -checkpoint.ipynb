```python
import cv2
import os
```

```python
loc='E:/ML_Codes/natural-images/images'
```

```python
import numpy as np
```

```python
labels=[]
for i in os.listdir(loc):
    if (i.split('_')[0])=='car':
        labels.append(0)
    elif (i.split('_')[0])=='motorbike':
        labels.append(1)
    elif (i.split('_')[0])=='person':
        labels.append(2)
```

```python
len(labels)
```

# https://pastebin.com/8JDUeH8B

```python
#Features
features=[]
for i in os.listdir(loc):
    path=os.path.join(loc,i)
    img=cv2.imread(path,0)
    img=cv2.resize(img,(100,100))
    features.append(img)
```

```python
ft=np.array(features)
ft=ft/255
lt=np.array(labels)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(ft,lt)
```

```python
nnet=tf.keras.models.Sequential()
nnet.add(tf.keras.layers.Flatten())
nnet.add(tf.keras.layers.Dense(128,activation='relu'))
nnet.add(tf.keras.layers.Dense(128,activation='relu'))
nnet.add(tf.keras.layers.Dense(3,activation='softmax'))
```

```python
nnet.compile(optimizer='adam',
            loss='sparse_categorical_crossentropy',
            metrics=['accuracy'])
```

```python
nnet.fit(xtrain,ytrain,epochs=5)
```

```python
pred=nnet.predict(xtest)
```
